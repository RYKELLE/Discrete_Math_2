$\textbf{Directed Graph}$


 is a type of graph where edges have a direction associated with them. Each edge in a directed graph connects a source vertex to a target vertex, indicating a one-way relationship or flow of information between the nodes. In my example, the edges (1, 2), (2, 3), (3, 4), and (4, 1) have a direction from the first node to the second node in each edge. This indicates a directed relationship or flow from the source node to the target node.

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.DiGraph()

G.add_nodes_from([1, 2, 3, 4])

G.add_edges_from([(1, 2), (2, 3), (3, 4), (4, 1)], color='blue')

pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True)


plt.show()
```

$\textbf{Simple Graphs}$


is an undirected graph where each edge connects two distinct vertices without self-loops or duplicate edges. In the simple graph, each of the nodes 1, 2, 3, and 4 represents a distinct entity, and the edges (1, 2), (2, 3), (3, 4), and (4, 1) represent connections between these entities. 

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()
G.add_nodes_from([1, 2, 3, 4])
G.add_edges_from([(1, 2), (2, 3), (3, 4), (4, 1)])

nx.draw(G, with_labels=True)

plt.show()
```

$\textbf{Complete Graphs}$


is a type of graph where every pair of distinct nodes (or vertices) is connected by a unique edge. In other words, in a complete graph, there is an edge between every pair of nodes. In this example each of the 10 nodes is connected to every other node by a single edge, resulting in a fully connected graph.

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.complete_graph(5)

nx.draw(G, with_labels=True)


plt.show()
```

$\textbf{Bipartite Graphs}$


A bipartite graph is a type of graph where the set of nodes can be divided into two disjoint sets such that no two nodes within the same set are connected by an edge. . In this example, nodes are divided into two distinct sets: set 1 (nodes 1, 2, and 3) and set 2 (nodes 'A', 'B', and 'C'), with edges only connecting nodes from different sets, visually represented with light blue and light green nodes, showcasing relationships between entities of different types.

```python
import networkx as nx
import matplotlib.pyplot as plt


G = nx.Graph()
G.add_nodes_from([1, 2, 3], bipartite=0) 
G.add_nodes_from(['A', 'B', 'C'], bipartite=1)
G.add_edges_from([(1, 'A'), (1, 'B'), (2, 'B'), (2, 'C'), (3, 'A')])

pos = nx.bipartite_layout(G, nodes=[1, 2, 3])  
nx.draw(G, pos, with_labels=True)

plt.show()
```

$\textbf{Subgraphs}$



A subgraph is a subset of the nodes and edges of a larger graph that forms a graph structure on its own. In this example the original graph is a simple undirected graph with nodes 1, 2, 3, 4, and 5 connected by edges, while the subgraph comprises nodes 1, 2, and 3 along with the edges connecting them, forming a connected graph structure within the selected subset.

```python
import networkx as nx
import matplotlib.pyplot as plt


G = nx.Graph()
G.add_nodes_from([1, 2, 3, 4, 5])
G.add_edges_from([(1, 2), (1, 3), (2, 3), (3, 4), (4, 5), (5, 1)])


sub_nodes = [1, 2, 3]
subgraph = G.subgraph(sub_nodes)


plt.figure(figsize=(10, 5))

plt.subplot(121)
nx.draw(G, with_labels=True, node_color='lightblue', edge_color='gray')
plt.title('Original Graph')


plt.subplot(122)
nx.draw(subgraph, with_labels=True, node_color='lightgreen', edge_color='black')
plt.title('Subgraph (Subset of Nodes 1, 2, 3)')

plt.tight_layout()
plt.show()
```

$\textbf{Degree of a Vertex}$

In this example the original graph is a simple undirected graph with nodes 1, 2, 3, 4, and 5 connected by edges, while the subgraph comprises nodes 1, 2, and 3 along with the edges connecting them, forming a connected graph structure within the selected subset.

$\textbf{Walk}$



a walk is a sequence of vertices and edges that starts at a vertex, follows edges from one vertex to another, and ends at a vertex. In the given graph, an example of a Walk could be $A \to B \to D \to E \to F$, meaning we start at node A, move to node B, then to node D, and so on, following the edges of the graph until we reach node F.

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()

G.add_nodes_from(['A', 'B', 'C', 'D', 'E', 'F'])
G.add_edges_from([('A', 'B'), ('A', 'C'), ('B', 'D'), ('C', 'D'), ('C', 'E'), ('D', 'E'), ('E', 'F')])

nx.draw(G, with_labels=True)
plt.show()
```

$\textbf{Trail}$

A trail is a walk in which all edges are distinct. An example of a trail in this graph could be: $Baguio \to Cagayan de Oro \to Davao \to Cebu \to Manila$. Note that in a trail, vertices may repeat, but edges must be unique.

```python
import networkx as nx
import matplotlib.pyplot as plt


G = nx.Graph()
cities = ['Manila', 'Cebu', 'Davao', 'Baguio', 'Iloilo', 'Cagayan de Oro']
G.add_nodes_from(cities)

connections = [('Manila', 'Cebu'), ('Manila', 'Davao'), ('Manila', 'Baguio'),
               ('Cebu', 'Davao'), ('Cebu', 'Iloilo'), ('Davao', 'Cagayan de Oro'),
               ('Baguio', 'Cagayan de Oro'), ('Iloilo', 'Cagayan de Oro')]

G.add_edges_from(connections)


nx.draw(G, with_labels=True)
plt.show()
```

$\textbf{Path}$

A "path" in graph theory is a trail in which all vertices are distinct. An example of a path in the example below could be: $1 \to 3 \to 6 \to 8 \to 5$:

```python
import networkx as nx
import matplotlib.pyplot as plt


G = nx.Graph()
nodes = [1, 2, 3, 4, 5, 6, 7, 8]
edges = [(1, 2), (1, 3), (2, 4), (2, 5), (3, 6), (3, 7), (4, 8), (5, 8), (6, 8), (7, 8)]
G.add_nodes_from(nodes)
G.add_edges_from(edges)


nx.draw(G, with_labels=True)
plt.show()
```

$\textbf{Closed Walk}$

A closed walk is a sequence of vertices and edges that starts and ends at the same vertex, while traversing edges without repeating any edge except possibly the first and last edge. An example of a closed walk in this graph could be: $A \to B \to C \to D \to E \to A$.

```python
import networkx as nx
import matplotlib.pyplot as plt


G = nx.Graph()
nodes = ['A', 'B', 'C', 'D', 'E']
edges = [('A', 'B'), ('B', 'C'), ('C', 'D'), ('D', 'E'), ('E', 'A')]

G.add_nodes_from(nodes)
G.add_edges_from(edges)

nx.draw(G, with_labels=True)
plt.show()
```

$\textbf{Circuit}$

A circuit is a closed walk in a graph where all edges are distinct except for the first and last edge, which are the same. In the graph below the cycle $X \to Y \to Z \to W \to X$ forms a circuit

```python
import networkx as nx
import matplotlib.pyplot as plt


G = nx.Graph()
nodes = ['X', 'Y', 'Z', 'W']
edges = [('X', 'Y'), ('Y', 'Z'), ('Z', 'W'), ('W', 'X'), ('X', 'Z')]

G.add_nodes_from(nodes)
G.add_edges_from(edges)


nx.draw(G, with_labels=True)
plt.show()

```

$\textbf{Simple Circuit}$

A simple circuit is a closed walk in a graph where all edges and vertices are distinct, except for the first and last vertex, which are the same. In the graph below, the cycle $P \to Q \to T \to S \to P$ forms a simple circuit.

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()
nodes = ['P', 'Q', 'R', 'S', 'T']
edges = [('P', 'Q'), ('Q', 'R'), ('R', 'S'), ('S', 'T'), ('T', 'P'), ('P', 'S'), ('Q', 'T')]

G.add_nodes_from(nodes)
G.add_edges_from(edges)

nx.draw(G, with_labels=True)
plt.show()
```

$\textbf{Connected Graph}$

A connected graph is a graph in which there is a path between every pair of vertices. This means that starting from any vertex, you can reach any other vertex by following edges in the graph. In the graph shown below, every node (A, B, C, D, E) is connected to every other node, either directly or through other nodes. This property of having a path between every pair of vertices makes the graph connected.

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()


nodes = ['A', 'B', 'C', 'D', 'E']
edges = [('A', 'B'), ('B', 'C'), ('C', 'D'), ('D', 'E')]

G.add_nodes_from(nodes)
G.add_edges_from(edges)

nx.draw(G, with_labels=True)
plt.show()
```

$\textbf{Euler's' Circuit}$

is a circuit in a graph that visits every edge exactly once and returns to the starting vertex. In other words, it is a closed walk that includes all edges of the graph. In the graph shown below, the circuit $A \to B \to C \to D \to A$ is an Eulerian circuit because it visits each edge (A-B, B-C, C-D, D-A) exactly once and returns to the starting vertex A.

```python
import networkx as nx
import matplotlib.pyplot as plt


G = nx.Graph()
nodes = ['A', 'B', 'C', 'D']
edges = [('A', 'B'), ('B', 'C'), ('C', 'D'), ('D', 'A')]

G.add_nodes_from(nodes)
G.add_edges_from(edges)


nx.draw(G, with_labels=True)
plt.show()
```

$\textbf{Euler's Trail}$

An Eulerian trail is a trail in a graph that visits every edge exactly once but may not return to the starting vertex. It is a path that includes all edges of the graph without repeating any edge. The Eulerian trail in this graph is $USC \to UC \to UV \to UP \to CIC \to USC \to UP$, which visits each edge exactly oncen without repeating any edge.

```python
import networkx as nx
import matplotlib.pyplot as plt


G = nx.Graph()
schools = ['USC', 'UC', 'UV', 'UP', 'CIC']
G.add_nodes_from(schools)

edges = [('USC', 'UC'), ('UC', 'UV'), ('UV', 'UP'), ('UP', 'CIC'), ('CIC', 'USC'), ('USC', 'UP')]
G.add_edges_from(edges)


nx.draw(G, with_labels=True)
plt.show()
```

$\textbf{Hamiltonian Circuit}$

A Hamiltonian circuit is a circuit in a graph that visits every vertex exactly once and returns to the starting vertex. In other words, it is a closed walk that includes all vertices of the graph. In the graph shown below, the circuit $A \to B \to C \to D \to E \to A$ is a Hamiltonian circuit because it visits each vertex (A, B, C, D, E) exactly once and returns to the starting vertex A.

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()
nodes = ['A', 'B', 'C', 'D', 'E']
G.add_nodes_from(nodes)


edges = [('A', 'B'), ('B', 'C'), ('C', 'D'), ('D', 'E'), ('E', 'A')]
G.add_edges_from(edges)

nx.draw(G, with_labels=True)

plt.show()
```
